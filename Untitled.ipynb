```python
# ============================================================
# Elegant fiscal time-series example figure:
# one normal KKT + several anomalous KKT series
#
# Reads:
# - final_ts_kkt_day_anomaly_ranking.csv
# - FINAL_COMBINED_KKT_ANOMALY_RANKING.csv
# - ts_kkt_id_okved_47.11_region_77_impute_True.csv
# - ts_kkt_id_okved_47.19_region_77_impute_True.csv
#
# Saves:
# - kkt_timeseries_normal_vs_anomalous.pdf
# - kkt_timeseries_normal_vs_anomalous.png
# - selected_kkt_for_timeseries_figure.csv
# ============================================================

from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ------------------------------------------------------------
# Paths
# ------------------------------------------------------------

DATA_ROOT = Path(r"C:\Users\ivan\WORK\workshops\ICDM\Ablation\kkt_data\test_pipeline_2025_2")
OUT_ROOT = Path(r"C:\Users\ivan\WORK\workshops\ICDM\Ablation\kkt_anomaly_outputs")
SUMMARY_DIR = OUT_ROOT / "summary"

TS_PATHS = {
    "47.11": DATA_ROOT / "dataloader_time_series" / "ts_kkt_id_okved_47.11_region_77_impute_True.csv",
    "47.19": DATA_ROOT / "dataloader_time_series" / "ts_kkt_id_okved_47.19_region_77_impute_True.csv",
}

DAY_RANKING_PATH = SUMMARY_DIR / "final_ts_kkt_day_anomaly_ranking.csv"
COMBINED_RANKING_PATH = SUMMARY_DIR / "FINAL_COMBINED_KKT_ANOMALY_RANKING.csv"

FIG_DIR = SUMMARY_DIR / "figures"
FIG_DIR.mkdir(parents=True, exist_ok=True)

FIG_PDF = FIG_DIR / "kkt_timeseries_normal_vs_anomalous.pdf"
FIG_PNG = FIG_DIR / "kkt_timeseries_normal_vs_anomalous.png"
SELECTED_CSV = FIG_DIR / "selected_kkt_for_timeseries_figure.csv"

print("DATA_ROOT:", DATA_ROOT)
print("SUMMARY_DIR:", SUMMARY_DIR)
print("FIG_DIR:", FIG_DIR)

for k, p in TS_PATHS.items():
    print(k, "exists:", p.exists(), p)

print("DAY_RANKING exists:", DAY_RANKING_PATH.exists())
print("COMBINED_RANKING exists:", COMBINED_RANKING_PATH.exists())

# ------------------------------------------------------------
# Config
# ------------------------------------------------------------

VALUE_COL = "total_sum"       # main fiscal time-series variable
AUX_COL = "n_bills"           # used for annotation / sanity
N_ANOMALOUS_SERIES = 4        # number of anomalous KKT devices to show
MIN_DAYS_FOR_SERIES = 40      # avoid too short lines
TOP_DAY_POOL = 300            # search anomalous KKT among top-N anomalous days
NORMAL_SAMPLE_POOL = 3000     # candidate normal devices sampled from non-top tail
RANDOM_STATE = 42

np.random.seed(RANDOM_STATE)

# ------------------------------------------------------------
# Load rankings
# ------------------------------------------------------------

day_rank = pd.read_csv(DAY_RANKING_PATH)
combined_rank = pd.read_csv(COMBINED_RANKING_PATH)

day_rank["date"] = pd.to_datetime(day_rank["date"], errors="coerce")
day_rank["kkt_id"] = day_rank["kkt_id"].astype(str)
combined_rank["kkt_id"] = combined_rank["kkt_id"].astype(str)

print("\nDay ranking:", day_rank.shape)
print("Combined ranking:", combined_rank.shape)

display(day_rank.head(10))
display(combined_rank.head(10))

# ------------------------------------------------------------
# Helper: decide which source file contains a KKT
# ------------------------------------------------------------

def get_ts_header(path: Path):
    return pd.read_csv(path, nrows=1).columns.tolist()

for okved, path in TS_PATHS.items():
    print(okved, get_ts_header(path)[:20])


def collect_rows_for_kkt_ids(path: Path, kkt_ids, usecols=None, chunksize=600_000):
    """
    Memory-safe loading of only selected KKT IDs from a large CSV.
    """
    kkt_ids = set(map(str, kkt_ids))
    chunks = []

    if usecols is None:
        header = pd.read_csv(path, nrows=1).columns.tolist()
        usecols = [c for c in ["kkt_id", "date", VALUE_COL, AUX_COL, "total_items", "ecash_fraction", "avg_sum"] if c in header]

    for chunk in pd.read_csv(path, usecols=usecols, chunksize=chunksize):
        chunk["kkt_id"] = chunk["kkt_id"].astype(str)
        sub = chunk[chunk["kkt_id"].isin(kkt_ids)].copy()
        if len(sub):
            chunks.append(sub)

    if not chunks:
        return pd.DataFrame(columns=usecols)

    out = pd.concat(chunks, ignore_index=True)
    out["date"] = pd.to_datetime(out["date"], errors="coerce")

    for c in out.columns:
        if c not in ["kkt_id", "date"]:
            out[c] = pd.to_numeric(out[c], errors="coerce")

    return out


def find_kkt_locations(kkt_ids):
    """
    Returns dict: kkt_id -> okved source.
    """
    remaining = set(map(str, kkt_ids))
    loc = {}

    for okved, path in TS_PATHS.items():
        if not remaining:
            break

        usecols = ["kkt_id"]
        found = set()

        for chunk in pd.read_csv(path, usecols=usecols, chunksize=700_000):
            ids = set(chunk["kkt_id"].astype(str).unique())
            hit = remaining & ids
            if hit:
                found |= hit

        for x in found:
            loc[x] = okved

        remaining -= found

    return loc


# ------------------------------------------------------------
# Select anomalous devices
# ------------------------------------------------------------
# We select KKT that:
# - appear in top anomalous KKT-day events
# - have repeated anomalous days if possible
# - are present in combined ranking
# ------------------------------------------------------------

top_days = day_rank.head(TOP_DAY_POOL).copy()

anom_device_stats = (
    top_days
    .groupby("kkt_id", as_index=False)
    .agg(
        n_top_days=("date", "nunique"),
        best_day_rank=("final_day_rank", "min") if "final_day_rank" in top_days.columns else ("final_day_anomaly_score", "idxmax"),
        best_day_score=("final_day_anomaly_score", "max"),
        max_total_sum=("total_sum", "max") if "total_sum" in top_days.columns else ("final_day_anomaly_score", "max"),
        median_total_sum=("total_sum", "median") if "total_sum" in top_days.columns else ("final_day_anomaly_score", "median"),
    )
)

if "final_kkt_anomaly_score" in combined_rank.columns:
    anom_device_stats = anom_device_stats.merge(
        combined_rank[["kkt_id", "final_kkt_anomaly_score"]].drop_duplicates("kkt_id"),
        on="kkt_id",
        how="left"
    )
else:
    anom_device_stats["final_kkt_anomaly_score"] = np.nan

anom_device_stats = anom_device_stats.sort_values(
    ["n_top_days", "best_day_score", "final_kkt_anomaly_score", "max_total_sum"],
    ascending=[False, False, False, False]
)

candidate_anom_ids = anom_device_stats["kkt_id"].head(20).tolist()
locs = find_kkt_locations(candidate_anom_ids)

selected_anom = []
for kkt in candidate_anom_ids:
    if kkt in locs:
        selected_anom.append(kkt)
    if len(selected_anom) >= N_ANOMALOUS_SERIES:
        break

print("\nSelected anomalous KKT:")
for k in selected_anom:
    print(k, "source OKVED:", locs.get(k))

# ------------------------------------------------------------
# Load anomalous series and filter by length
# ------------------------------------------------------------

all_selected_ids = list(selected_anom)
anom_frames = []

for okved, path in TS_PATHS.items():
    ids_here = [k for k in selected_anom if locs.get(k) == okved]
    if ids_here:
        df_part = collect_rows_for_kkt_ids(path, ids_here)
        df_part["_okved_source"] = okved
        anom_frames.append(df_part)

anom_ts = pd.concat(anom_frames, ignore_index=True) if anom_frames else pd.DataFrame()

# Keep only devices with enough days
valid_anom = []
for kkt, g in anom_ts.groupby("kkt_id"):
    if g["date"].nunique() >= MIN_DAYS_FOR_SERIES:
        valid_anom.append(kkt)

selected_anom = valid_anom[:N_ANOMALOUS_SERIES]
print("\nValid anomalous KKT after length filter:", selected_anom)

if len(selected_anom) == 0:
    raise RuntimeError("No anomalous KKT series found with enough days. Lower MIN_DAYS_FOR_SERIES.")

# ------------------------------------------------------------
# Select one normal device
# ------------------------------------------------------------
# Normal candidate:
# - low combined anomaly score
# - not in top KKT-day anomaly ranking
# - enough days
# ------------------------------------------------------------

top_anom_ids_global = set(day_rank.head(5000)["kkt_id"].astype(str).unique())
combined_sorted = combined_rank.copy()

score_col = "final_kkt_anomaly_score" if "final_kkt_anomaly_score" in combined_sorted.columns else None

if score_col:
    combined_sorted = combined_sorted.sort_values(score_col, ascending=True)

normal_candidates = combined_sorted[~combined_sorted["kkt_id"].isin(top_anom_ids_global)]["kkt_id"].astype(str).head(NORMAL_SAMPLE_POOL).tolist()

normal_locs = find_kkt_locations(normal_candidates[:800])  # enough for search

normal_id = None
normal_okved = None
normal_ts = pd.DataFrame()

for okved, path in TS_PATHS.items():
    ids_here = [k for k in normal_candidates if normal_locs.get(k) == okved]
    if not ids_here:
        continue

    # load in batches of candidate IDs to avoid reading too much repeatedly
    for start in range(0, len(ids_here), 100):
        batch = ids_here[start:start+100]
        tmp = collect_rows_for_kkt_ids(path, batch)
        if tmp.empty:
            continue

        for kkt, g in tmp.groupby("kkt_id"):
            g = g.sort_values("date")
            if g["date"].nunique() >= MIN_DAYS_FOR_SERIES:
                # avoid flat zero series
                if g[VALUE_COL].fillna(0).sum() > 0 and g[VALUE_COL].std(skipna=True) > 0:
                    normal_id = kkt
                    normal_okved = okved
                    normal_ts = g.copy()
                    break

        if normal_id is not None:
            break

    if normal_id is not None:
        break

if normal_id is None:
    raise RuntimeError("Could not find a normal KKT with enough days. Increase NORMAL_SAMPLE_POOL.")

print("\nSelected normal KKT:", normal_id, "source OKVED:", normal_okved)

# ------------------------------------------------------------
# Build final plotting dataframe
# ------------------------------------------------------------

selected_ids = [normal_id] + selected_anom
selected_locs = find_kkt_locations(selected_ids)

frames = []
for okved, path in TS_PATHS.items():
    ids_here = [k for k in selected_ids if selected_locs.get(k) == okved]
    if ids_here:
        part = collect_rows_for_kkt_ids(path, ids_here)
        part["_okved_source"] = okved
        frames.append(part)

plot_df = pd.concat(frames, ignore_index=True)
plot_df = plot_df[plot_df["kkt_id"].isin(selected_ids)].copy()
plot_df["date"] = pd.to_datetime(plot_df["date"], errors="coerce")
plot_df = plot_df.sort_values(["kkt_id", "date"])

# Join day anomaly scores for highlighted points
day_scores = day_rank[["kkt_id", "date", "final_day_anomaly_score"]].copy()
day_scores["kkt_id"] = day_scores["kkt_id"].astype(str)
day_scores["date"] = pd.to_datetime(day_scores["date"], errors="coerce")

plot_df = plot_df.merge(day_scores, on=["kkt_id", "date"], how="left")
plot_df["final_day_anomaly_score"] = plot_df["final_day_anomaly_score"].fillna(0.0)

# Create shorter labels
label_map = {normal_id: "Normal KKT"}
for i, k in enumerate(selected_anom, start=1):
    label_map[k] = f"Anomalous KKT {i}"

plot_df["series_label"] = plot_df["kkt_id"].map(label_map)

# Save selected info
selected_info = []
for k in selected_ids:
    selected_info.append({
        "kkt_id": k,
        "label": label_map.get(k),
        "okved_source": selected_locs.get(k),
        "n_days": int(plot_df.loc[plot_df["kkt_id"] == k, "date"].nunique()),
        "sum_total_sum": float(plot_df.loc[plot_df["kkt_id"] == k, VALUE_COL].sum()),
        "max_total_sum": float(plot_df.loc[plot_df["kkt_id"] == k, VALUE_COL].max()),
        "n_highlighted_anomaly_days": int((plot_df.loc[plot_df["kkt_id"] == k, "final_day_anomaly_score"] > 0).sum()),
    })

selected_info_df = pd.DataFrame(selected_info)
selected_info_df.to_csv(SELECTED_CSV, index=False, encoding="utf-8-sig")
display(selected_info_df)

# ------------------------------------------------------------
# Plot
# ------------------------------------------------------------

# Use log scale to make normal and anomalous lines comparable.
# Create y value.
plot_df["plot_value"] = pd.to_numeric(plot_df[VALUE_COL], errors="coerce").fillna(0.0)
plot_df["plot_value"] = plot_df["plot_value"].clip(lower=0)

# Highlight only strongest anomalous days from selected series.
# Threshold can be changed.
HIGHLIGHT_SCORE_THRESHOLD = 0.95

fig, axes = plt.subplots(
    nrows=len(selected_ids),
    ncols=1,
    figsize=(10.5, 2.0 + 1.65 * len(selected_ids)),
    sharex=True,
    constrained_layout=True
)

if len(selected_ids) == 1:
    axes = [axes]

for ax, kkt in zip(axes, selected_ids):
    g = plot_df[plot_df["kkt_id"] == kkt].sort_values("date").copy()
    label = label_map[kkt]

    # Main line
    ax.plot(
        g["date"],
        g["plot_value"],
        linewidth=1.6,
        marker="o",
        markersize=2.5,
        alpha=0.9,
        label=label
    )

    # Rolling median for visual baseline
    g_roll = g.set_index("date")["plot_value"].rolling(7, min_periods=3).median()
    ax.plot(
        g_roll.index,
        g_roll.values,
        linewidth=1.1,
        linestyle="--",
        alpha=0.8,
        label="7-day rolling median"
    )

    # Highlight anomalous days
    gh = g[g["final_day_anomaly_score"] >= HIGHLIGHT_SCORE_THRESHOLD].copy()
    if len(gh):
        ax.scatter(
            gh["date"],
            gh["plot_value"],
            s=42,
            marker="x",
            linewidths=1.6,
            zorder=5,
            label="ranked anomalous day"
        )

    # Title and annotations
    max_val = g["plot_value"].max()
    n_days = g["date"].nunique()
    n_anom_days = int((g["final_day_anomaly_score"] >= HIGHLIGHT_SCORE_THRESHOLD).sum())

    ax.set_title(
        f"{label} | days={n_days}, max daily turnover={max_val:,.0f}, highlighted days={n_anom_days}",
        loc="left",
        fontsize=9
    )

    ax.set_ylabel("Daily turnover")
    ax.grid(True, linewidth=0.4, alpha=0.35)

    # Log scale helps if anomalous values are huge.
    # symlog keeps zero visible.
    ax.set_yscale("symlog", linthresh=1000)

    # Make y tick labels compact
    ax.tick_params(axis="both", labelsize=8)

    # Legend only for first subplot to avoid clutter
    if ax == axes[0]:
        ax.legend(loc="upper right", fontsize=8, frameon=True)

axes[-1].set_xlabel("Date")
axes[-1].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
axes[-1].xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))

fig.suptitle(
    "Examples of normal and anomalous fiscal time series in KKT data",
    fontsize=12,
    y=1.01
)

fig.savefig(FIG_PDF, bbox_inches="tight")
fig.savefig(FIG_PNG, dpi=300, bbox_inches="tight")

print("\nSaved:")
print("PDF:", FIG_PDF)
print("PNG:", FIG_PNG)
print("Selected:", SELECTED_CSV)

plt.show()
```

```python
# ------------------------------------------------------------
# Plot: normalized values and DD.MM time axis
# ------------------------------------------------------------

# Raw value
plot_df["raw_value"] = pd.to_numeric(plot_df[VALUE_COL], errors="coerce").fillna(0.0)
plot_df["raw_value"] = plot_df["raw_value"].clip(lower=0)

# Per-series normalization:
# value_a.u. = daily value / median positive daily value for this KKT.
# This keeps the shape and avoids reporting absolute fiscal sums.
plot_df["plot_value_au"] = np.nan

norm_rows = []
for kkt, g in plot_df.groupby("kkt_id"):
    positive = g.loc[g["raw_value"] > 0, "raw_value"]
    if len(positive) > 0:
        scale = float(positive.median())
    else:
        scale = 1.0

    if not np.isfinite(scale) or scale <= 0:
        scale = 1.0

    idx = plot_df["kkt_id"] == kkt
    plot_df.loc[idx, "plot_value_au"] = plot_df.loc[idx, "raw_value"] / scale

    norm_rows.append({
        "kkt_id": kkt,
        "label": label_map.get(kkt, kkt),
        "normalization": "raw daily turnover divided by median positive daily turnover of the same KKT",
        "scale_value_hidden": scale,
        "n_days": int(g["date"].nunique()),
    })

norm_df = pd.DataFrame(norm_rows)
norm_df.to_csv(FIG_DIR / "timeseries_normalization_info.csv", index=False, encoding="utf-8-sig")

# Highlight only strongest anomalous days from selected series.
HIGHLIGHT_SCORE_THRESHOLD = 0.95

fig, axes = plt.subplots(
    nrows=len(selected_ids),
    ncols=1,
    figsize=(10.5, 2.0 + 1.55 * len(selected_ids)),
    sharex=True,
    constrained_layout=True
)

if len(selected_ids) == 1:
    axes = [axes]

for ax, kkt in zip(axes, selected_ids):
    g = plot_df[plot_df["kkt_id"] == kkt].sort_values("date").copy()
    label = label_map[kkt]

    # Main line
    ax.plot(
        g["date"],
        g["plot_value_au"],
        linewidth=1.5,
        marker="o",
        markersize=2.3,
        alpha=0.9,
        label=label
    )

    # Rolling median in normalized units
    g_roll = g.set_index("date")["plot_value_au"].rolling(7, min_periods=3).median()
    ax.plot(
        g_roll.index,
        g_roll.values,
        linewidth=1.05,
        linestyle="--",
        alpha=0.85,
        label="7-day rolling median"
    )

    # Highlight anomalous days
    gh = g[g["final_day_anomaly_score"] >= HIGHLIGHT_SCORE_THRESHOLD].copy()
    if len(gh):
        ax.scatter(
            gh["date"],
            gh["plot_value_au"],
            s=40,
            marker="x",
            linewidths=1.5,
            zorder=5,
            label="ranked anomalous day"
        )

    n_days = g["date"].nunique()
    n_anom_days = int((g["final_day_anomaly_score"] >= HIGHLIGHT_SCORE_THRESHOLD).sum())

    ax.set_title(
        f"{label} | days={n_days}, highlighted days={n_anom_days}",
        loc="left",
        fontsize=9
    )

    ax.set_ylabel("Normalized\nturnover, a.u.")
    ax.grid(True, linewidth=0.4, alpha=0.35)

    # Log-like scale with zeros allowed.
    # In normalized units, this gives readable dynamics without exposing sums.
    ax.set_yscale("symlog", linthresh=0.05)

    ax.tick_params(axis="both", labelsize=8)

    if ax == axes[0]:
        ax.legend(loc="upper right", fontsize=8, frameon=True)

axes[-1].set_xlabel("Date")

# DD.MM format, no year
axes[-1].xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
axes[-1].xaxis.set_major_formatter(mdates.DateFormatter("%d.%m"))

# Rotate dates slightly for readability
for label in axes[-1].get_xticklabels():
    label.set_rotation(0)
    label.set_horizontalalignment("center")

fig.suptitle(
    "Examples of normal and anomalous fiscal time series",
    fontsize=12,
    y=1.01
)

FIG_PDF = FIG_DIR / "kkt_timeseries_normal_vs_anomalous_normalized.pdf"
FIG_PNG = FIG_DIR / "kkt_timeseries_normal_vs_anomalous_normalized.png"

fig.savefig(FIG_PDF, bbox_inches="tight")
fig.savefig(FIG_PNG, dpi=300, bbox_inches="tight")

print("\nSaved:")
print("PDF:", FIG_PDF)
print("PNG:", FIG_PNG)
print("Normalization info:", FIG_DIR / "timeseries_normalization_info.csv")

plt.show()
```
