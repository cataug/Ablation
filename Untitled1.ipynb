````python
from pathlib import Path
from collections import Counter, defaultdict
import os
import json
import csv
from datetime import datetime

# =============================================================================
# CONFIG
# =============================================================================

ROOT = Path(r"C:\Users\ivan\WORK\workshops\ICDM\Ablation")

OUT_REPORT_DIR = ROOT / "_structure_report"

MAX_TREE_DEPTH = 5
MAX_ITEMS_PER_DIR = 40
MAX_EXAMPLES_PER_EXT = 15
TOP_N_LARGEST_FILES = 50
TOP_N_DIRS = 50


# =============================================================================
# HELPERS
# =============================================================================

def human_size(num_bytes: int) -> str:
    units = ["B", "KB", "MB", "GB", "TB", "PB"]
    size = float(num_bytes)

    for unit in units:
        if size < 1024:
            return f"{size:.2f} {unit}"
        size /= 1024

    return f"{size:.2f} EB"


def safe_stat(path: Path):
    try:
        return path.stat()
    except Exception:
        return None


def get_depth(path: Path, root: Path) -> int:
    try:
        return len(path.relative_to(root).parts)
    except Exception:
        return 0


def is_probably_heavy_or_generated(path: Path) -> bool:
    ext = path.suffix.lower()
    heavy_exts = {
        ".pt", ".pth", ".ckpt", ".bin",
        ".npy", ".npz",
        ".pkl", ".pickle",
        ".gz", ".zip", ".tar", ".rar", ".7z",
        ".parquet", ".feather",
        ".h5", ".hdf5",
    }

    parts = {p.lower() for p in path.parts}
    generated_parts = {
        "__pycache__",
        ".ipynb_checkpoints",
        "wandb",
        "runs",
        "outputs",
        "checkpoints",
        "models",
        "data",
        "raw",
        "processed",
    }

    return ext in heavy_exts or bool(parts & generated_parts)


def repo_candidate_score(path: Path, root: Path) -> tuple[int, str]:
    """
    Lower score = more useful for GitHub.
    """
    rel = path.relative_to(root)
    name = path.name.lower()
    ext = path.suffix.lower()
    parts = [p.lower() for p in rel.parts]

    if ".ipynb_checkpoints" in parts or "__pycache__" in parts:
        return 999, "exclude cache/checkpoint"

    if ext in {".pt", ".pth", ".ckpt", ".bin", ".npy", ".npz", ".pkl", ".pickle", ".gz", ".zip", ".tar", ".7z", ".rar"}:
        return 900, "exclude heavy binary/archive"

    if any(p in {"data", "raw", "processed", "checkpoints", "models"} for p in parts):
        return 850, "probably generated data/model artifact"

    if ext == ".py":
        return 10, "high-priority code"

    if ext == ".ipynb":
        return 20, "notebook"

    if ext in {".md", ".txt", ".yaml", ".yml", ".toml", ".json"}:
        return 30, "config/report/documentation"

    if ext in {".csv", ".xlsx"}:
        if "result" in name or "summary" in name or "ablation" in name or "metric" in name:
            return 40, "result table"
        return 55, "tabular file"

    if ext in {".pdf", ".png", ".jpg", ".jpeg", ".svg"}:
        if "plot" in parts or "fig" in name or "figure" in name or "ablation" in name:
            return 60, "figure/plot"
        return 75, "image/pdf"

    if ext in {".log"}:
        return 70, "log file"

    return 100, "other lightweight file"


# =============================================================================
# SCAN
# =============================================================================

def scan_structure(root: Path):
    if not root.exists():
        raise FileNotFoundError(f"Root folder does not exist: {root}")

    if not root.is_dir():
        raise NotADirectoryError(f"Path is not a directory: {root}")

    total_files = 0
    total_dirs = 0
    total_size = 0

    ext_counter = Counter()
    depth_counter = Counter()
    dir_file_counts = Counter()
    dir_total_size = Counter()
    examples_by_ext = defaultdict(list)

    largest_files = []
    repo_candidates = []
    suspicious_heavy = []

    for current_dir, dirnames, filenames in os.walk(root):
        current_path = Path(current_dir)
        total_dirs += 1

        for filename in filenames:
            file_path = current_path / filename

            st = safe_stat(file_path)
            size = st.st_size if st else 0

            total_files += 1
            total_size += size

            ext = file_path.suffix.lower() or "[no extension]"

            rel_depth = get_depth(file_path, root)
            depth_counter[rel_depth] += 1
            ext_counter[ext] += 1

            dir_file_counts[current_path] += 1
            dir_total_size[current_path] += size

            if len(examples_by_ext[ext]) < MAX_EXAMPLES_PER_EXT:
                examples_by_ext[ext].append(file_path)

            largest_files.append((size, file_path))

            score, reason = repo_candidate_score(file_path, root)
            repo_candidates.append((score, reason, size, file_path))

            if is_probably_heavy_or_generated(file_path) or size > 25 * 1024 * 1024:
                suspicious_heavy.append((size, file_path))

    largest_files.sort(reverse=True, key=lambda x: x[0])
    suspicious_heavy.sort(reverse=True, key=lambda x: x[0])
    repo_candidates.sort(key=lambda x: (x[0], str(x[3]).lower()))

    return {
        "total_files": total_files,
        "total_dirs": total_dirs,
        "total_size": total_size,
        "ext_counter": ext_counter,
        "depth_counter": depth_counter,
        "dir_file_counts": dir_file_counts,
        "dir_total_size": dir_total_size,
        "examples_by_ext": examples_by_ext,
        "largest_files": largest_files,
        "suspicious_heavy": suspicious_heavy,
        "repo_candidates": repo_candidates,
    }


# =============================================================================
# PRINT TREE
# =============================================================================

def print_tree(root: Path, max_depth: int = 5, max_items: int = 40):
    print("\n" + "=" * 120)
    print("DIRECTORY TREE")
    print("=" * 120)

    def _print_dir(current: Path, depth: int):
        if depth > max_depth:
            return

        indent = "    " * depth
        name = current.name if depth > 0 else str(current)
        print(f"{indent}{name}/")

        try:
            items = sorted(
                current.iterdir(),
                key=lambda p: (not p.is_dir(), p.name.lower())
            )
        except Exception as e:
            print(f"{indent}    [ERROR reading directory: {e}]")
            return

        shown = items[:max_items]
        hidden = len(items) - len(shown)

        for item in shown:
            if item.is_dir():
                _print_dir(item, depth + 1)
            else:
                st = safe_stat(item)
                size = human_size(st.st_size) if st else "?"
                print(f"{indent}    {item.name}  [{size}]")

        if hidden > 0:
            print(f"{indent}    ... {hidden} more items not shown")

    _print_dir(root, 0)


# =============================================================================
# PRINT SUMMARY
# =============================================================================

def print_summary(result):
    print("\n" + "=" * 120)
    print("SUMMARY")
    print("=" * 120)

    print(f"Root              : {ROOT}")
    print(f"Total directories : {result['total_dirs']}")
    print(f"Total files       : {result['total_files']}")
    print(f"Total size        : {human_size(result['total_size'])}")

    print("\n" + "=" * 120)
    print("FILES BY EXTENSION")
    print("=" * 120)

    for ext, count in result["ext_counter"].most_common():
        print(f"{ext:20s} {count:10d}")

    print("\n" + "=" * 120)
    print("FILES BY DEPTH")
    print("=" * 120)

    for depth, count in sorted(result["depth_counter"].items()):
        print(f"Depth {depth:2d}: {count:10d} files")

    print("\n" + "=" * 120)
    print("TOP DIRECTORIES BY FILE COUNT")
    print("=" * 120)

    for path, count in result["dir_file_counts"].most_common(TOP_N_DIRS):
        rel = path.relative_to(ROOT)
        size = human_size(result["dir_total_size"][path])
        print(f"{str(rel):90s} files={count:8d} size={size}")

    print("\n" + "=" * 120)
    print("TOP DIRECTORIES BY SIZE")
    print("=" * 120)

    for path, size in result["dir_total_size"].most_common(TOP_N_DIRS):
        rel = path.relative_to(ROOT)
        count = result["dir_file_counts"][path]
        print(f"{str(rel):90s} size={human_size(size):>12s} files={count:8d}")

    print("\n" + "=" * 120)
    print("LARGEST FILES")
    print("=" * 120)

    for size, path in result["largest_files"][:TOP_N_LARGEST_FILES]:
        rel = path.relative_to(ROOT)
        print(f"{human_size(size):>12s}  {rel}")

    print("\n" + "=" * 120)
    print("PROBABLY HEAVY / GENERATED FILES")
    print("=" * 120)

    if result["suspicious_heavy"]:
        for size, path in result["suspicious_heavy"][:TOP_N_LARGEST_FILES]:
            rel = path.relative_to(ROOT)
            print(f"{human_size(size):>12s}  {rel}")
    else:
        print("No obvious heavy/generated files found.")

    print("\n" + "=" * 120)
    print("EXAMPLES BY EXTENSION")
    print("=" * 120)

    for ext, examples in sorted(result["examples_by_ext"].items()):
        print(f"\n[{ext}]")
        for p in examples:
            print(f"  {p.relative_to(ROOT)}")

    print("\n" + "=" * 120)
    print("TOP GITHUB CANDIDATES")
    print("=" * 120)

    shown = 0
    for score, reason, size, path in result["repo_candidates"]:
        if score >= 900:
            continue

        rel = path.relative_to(ROOT)
        print(f"[{score:3d}] {human_size(size):>10s}  {rel}  -- {reason}")
        shown += 1

        if shown >= 120:
            break


# =============================================================================
# SAVE REPORTS
# =============================================================================

def save_reports(result):
    OUT_REPORT_DIR.mkdir(parents=True, exist_ok=True)

    summary = {
        "created_at": datetime.now().isoformat(timespec="seconds"),
        "root": str(ROOT),
        "total_directories": result["total_dirs"],
        "total_files": result["total_files"],
        "total_size_bytes": result["total_size"],
        "total_size_human": human_size(result["total_size"]),
        "files_by_extension": dict(result["ext_counter"].most_common()),
        "files_by_depth": dict(sorted(result["depth_counter"].items())),
        "largest_files": [
            {
                "path": str(path.relative_to(ROOT)),
                "size_bytes": size,
                "size_human": human_size(size),
            }
            for size, path in result["largest_files"][:TOP_N_LARGEST_FILES]
        ],
        "probably_heavy_or_generated_files": [
            {
                "path": str(path.relative_to(ROOT)),
                "size_bytes": size,
                "size_human": human_size(size),
            }
            for size, path in result["suspicious_heavy"][:TOP_N_LARGEST_FILES]
        ],
        "top_github_candidates": [
            {
                "path": str(path.relative_to(ROOT)),
                "size_bytes": size,
                "size_human": human_size(size),
                "score": score,
                "reason": reason,
            }
            for score, reason, size, path in result["repo_candidates"][:300]
            if score < 900
        ],
    }

    json_path = OUT_REPORT_DIR / "ablation_structure_summary.json"
    json_path.write_text(json.dumps(summary, indent=2, ensure_ascii=False), encoding="utf-8")

    csv_path = OUT_REPORT_DIR / "ablation_github_candidates.csv"
    with open(csv_path, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(
            f,
            fieldnames=["score", "reason", "size_bytes", "size_human", "path"]
        )
        writer.writeheader()

        for score, reason, size, path in result["repo_candidates"]:
            writer.writerow({
                "score": score,
                "reason": reason,
                "size_bytes": size,
                "size_human": human_size(size),
                "path": str(path.relative_to(ROOT)),
            })

    all_files_csv = OUT_REPORT_DIR / "ablation_all_files.csv"
    with open(all_files_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(
            f,
            fieldnames=["size_bytes", "size_human", "extension", "path"]
        )
        writer.writeheader()

        for size, path in result["largest_files"]:
            writer.writerow({
                "size_bytes": size,
                "size_human": human_size(size),
                "extension": path.suffix.lower() or "[no extension]",
                "path": str(path.relative_to(ROOT)),
            })

    print("\n" + "=" * 120)
    print("SAVED REPORTS")
    print("=" * 120)
    print(json_path)
    print(csv_path)
    print(all_files_csv)


# =============================================================================
# MAIN
# =============================================================================

def main():
    result = scan_structure(ROOT)
    print_summary(result)
    print_tree(ROOT, max_depth=MAX_TREE_DEPTH, max_items=MAX_ITEMS_PER_DIR)
    save_reports(result)

    print("\n" + "=" * 120)
    print("DONE")
    print("=" * 120)


if __name__ == "__main__":
    main()
````

````python
from pathlib import Path
import shutil
import json
import csv
from datetime import datetime

# =============================================================================
# CONFIG
# =============================================================================

SRC_ROOT = Path(r"C:\Users\ivan\WORK\workshops\ICDM\Ablation")
DST_ROOT = Path(r"C:\Users\ivan\WORK\workshops\ICDM\Ablation-github")

PROJECT_NAME = "Graph-Anomaly-Ablation"

MAX_FILES = 100
MAX_FILE_MB = 25

INCLUDE_NOTEBOOKS = True
INCLUDE_GADBENCH_CODE = True
INCLUDE_SUMMARY_FIGURES = True
INCLUDE_SELECTED_JSON = True
INCLUDE_SELECTED_LOGS = True

# =============================================================================
# HARD EXCLUDES
# =============================================================================

EXCLUDE_TOP_DIRS = {
    ".git",
    ".ipynb_checkpoints",
    "__pycache__",
    "kkt_data",
}

EXCLUDE_DIR_PARTS = {
    ".git",
    ".ipynb_checkpoints",
    "__pycache__",
    ".pytest_cache",
    ".mypy_cache",
    ".ruff_cache",
    "cache",
    "cached_features",
    "cached_perturbed_features",
    "raw",
    "processed",
    "dgl_raw",
    "dgl_processed",
    "pyg",
    "exports",
    "models_cache",
    "checkpoints",
}

EXCLUDE_EXTENSIONS = {
    ".pt",
    ".pth",
    ".ckpt",
    ".bin",
    ".npz",
    ".npy",
    ".pkl",
    ".pickle",
    ".gz",
    ".zip",
    ".tar",
    ".7z",
    ".rar",
    ".mat",
    ".h5",
    ".hdf5",
    ".parquet",
    ".feather",
}

# Huge generated CSVs should not be copied even if CSV is normally allowed.
EXCLUDE_NAME_CONTAINS = {
    "__ranked_kkt_days",
    "__ranked.csv",
    "ts_features",
}

# =============================================================================
# README / GITIGNORE
# =============================================================================

README_LINES = [
    "# Graph Anomaly Ablation Experiments",
    "",
    "This repository contains a lightweight GitHub snapshot of graph anomaly detection ablation experiments.",
    "It includes experiment notebooks, selected source code, compact summaries, logs, result tables, and figures.",
    "",
    "Large datasets, processed graph tensors, cached feature matrices, and large ranked output tables are intentionally excluded.",
    "",
    "## Repository contents",
    "",
    "```text",
    "*.ipynb",
    "    Main experiment and analysis notebooks.",
    "",
    "graph_anomaly_datasets/repos/GADBench-master/",
    "    Selected lightweight reference scripts from the local GADBench copy, if present.",
    "",
    "ablation_trivial_baselines_outputs/summary/",
    "    Aggregated baseline ablation summaries.",
    "",
    "kkt_anomaly_outputs/summary/",
    "    Compact KKT anomaly detection summaries and selected figures.",
    "",
    "kkt_data_audit_outputs/",
    "    Dataset audit reports and lightweight samples.",
    "",
    "robustness_attribute_structure_outputs/summary/",
    "    Robustness summary outputs.",
    "",
    "MANIFEST.json",
    "    File selection manifest.",
    "```",
    "",
    "## Excluded files",
    "",
    "The original working directory contains large generated artifacts and datasets. These are excluded from GitHub:",
    "",
    "```text",
    "kkt_data/",
    "kkt_anomaly_outputs/cache/",
    "kkt_anomaly_outputs/results_csv/",
    "graph_anomaly_datasets/pyg/",
    "graph_anomaly_datasets/dgl_raw/",
    "graph_anomaly_datasets/dgl_processed/",
    "graph_anomaly_datasets/exports/",
    "*.pt",
    "*.npz",
    "*.bin",
    "*.gz",
    "*.zip",
    "*.mat",
    "```",
    "",
    "## Notes",
    "",
    "The repository is intended to preserve the reproducible experimental logic and compact result evidence without storing raw or processed datasets.",
    "",
]

README_MD = "\n".join(README_LINES) + "\n"

GITIGNORE_LINES = [
    "# Python",
    "__pycache__/",
    "*.py[cod]",
    ".pytest_cache/",
    ".mypy_cache/",
    ".ruff_cache/",
    "",
    "# Jupyter",
    ".ipynb_checkpoints/",
    "",
    "# OS / IDE",
    ".DS_Store",
    "Thumbs.db",
    ".vscode/",
    ".idea/",
    "",
    "# Data and generated heavy outputs",
    "kkt_data/",
    "graph_anomaly_datasets/pyg/",
    "graph_anomaly_datasets/dgl_raw/",
    "graph_anomaly_datasets/dgl_processed/",
    "graph_anomaly_datasets/exports/",
    "**/raw/",
    "**/processed/",
    "**/cache/",
    "**/cached_features/",
    "**/cached_perturbed_features/",
    "",
    "# Heavy formats",
    "*.pt",
    "*.pth",
    "*.ckpt",
    "*.bin",
    "*.npz",
    "*.npy",
    "*.pkl",
    "*.pickle",
    "*.gz",
    "*.zip",
    "*.tar",
    "*.7z",
    "*.rar",
    "*.mat",
    "*.h5",
    "*.hdf5",
    "*.parquet",
    "*.feather",
    "",
    "# Huge generated tables",
    "*__ranked_kkt_days.csv",
    "*__ranked.csv",
    "*ts_features.csv",
    "",
]

GITIGNORE = "\n".join(GITIGNORE_LINES) + "\n"

# =============================================================================
# HELPERS
# =============================================================================

def human_size(n: int) -> str:
    units = ["B", "KB", "MB", "GB", "TB"]
    x = float(n)
    for u in units:
        if x < 1024:
            return f"{x:.2f} {u}"
        x /= 1024
    return f"{x:.2f} PB"


def safe_size(path: Path) -> int:
    try:
        return path.stat().st_size
    except OSError:
        return 0


def is_valid_file(path: Path) -> bool:
    if not path.is_file():
        return False

    try:
        rel = path.relative_to(SRC_ROOT)
    except ValueError:
        return False

    if not rel.parts:
        return False

    top = rel.parts[0]
    parts_lower = {p.lower() for p in rel.parts}
    name_lower = path.name.lower()

    if top in EXCLUDE_TOP_DIRS:
        return False

    if parts_lower & EXCLUDE_DIR_PARTS:
        return False

    if path.suffix.lower() in EXCLUDE_EXTENSIONS:
        return False

    if any(token.lower() in name_lower for token in EXCLUDE_NAME_CONTAINS):
        return False

    if safe_size(path) > MAX_FILE_MB * 1024 * 1024:
        return False

    return True


def add_candidate(candidates, path: Path, priority: int, reason: str):
    if not path.exists() or not is_valid_file(path):
        return

    rel = path.relative_to(SRC_ROOT)

    candidates.append({
        "src": path,
        "rel": rel,
        "priority": priority,
        "reason": reason,
        "size": safe_size(path),
        "sort_key": str(rel).lower(),
    })


def add_exact(candidates, rel_path: str, priority: int, reason: str):
    add_candidate(candidates, SRC_ROOT / rel_path, priority, reason)


def add_glob(candidates, pattern: str, priority: int, reason: str, limit=None):
    files = [p for p in SRC_ROOT.glob(pattern) if is_valid_file(p)]
    files = sorted(files, key=lambda p: str(p.relative_to(SRC_ROOT)).lower())

    if limit is not None:
        files = files[:limit]

    for p in files:
        add_candidate(candidates, p, priority, reason)


def deduplicate(candidates):
    seen = set()
    out = []

    for item in candidates:
        key = str(item["rel"]).lower()
        if key in seen:
            continue
        seen.add(key)
        out.append(item)

    return out


def copy_file(src: Path, dst: Path):
    dst.parent.mkdir(parents=True, exist_ok=True)
    shutil.copy2(src, dst)


def write_text(path: Path, text: str):
    path.parent.mkdir(parents=True, exist_ok=True)
    path.write_text(text, encoding="utf-8")


# =============================================================================
# CANDIDATE SELECTION
# =============================================================================

def build_candidates():
    candidates = []

    # -------------------------------------------------------------------------
    # 1. Main notebooks
    # -------------------------------------------------------------------------
    if INCLUDE_NOTEBOOKS:
        # Only top-level notebooks, not checkpoints.
        add_glob(candidates, "*.ipynb", 1, "main top-level notebook")

    # -------------------------------------------------------------------------
    # 2. Own top-level code / docs, if present
    # -------------------------------------------------------------------------
    add_glob(candidates, "*.py", 5, "top-level Python script")
    add_glob(candidates, "*.md", 6, "top-level markdown documentation")
    add_glob(candidates, "*.txt", 7, "top-level text documentation")

    # -------------------------------------------------------------------------
    # 3. Selected GADBench lightweight code only; no datasets
    # -------------------------------------------------------------------------
    if INCLUDE_GADBENCH_CODE:
        add_glob(
            candidates,
            r"graph_anomaly_datasets/repos/GADBench-master/*.py",
            10,
            "selected GADBench lightweight script",
        )
        add_glob(
            candidates,
            r"graph_anomaly_datasets/repos/GADBench-master/models/*.py",
            11,
            "selected GADBench model code",
        )
        add_exact(
            candidates,
            r"graph_anomaly_datasets\repos\GADBench-master\readme.md",
            12,
            "GADBench readme",
        )

    # -------------------------------------------------------------------------
    # 4. Most important aggregate summaries
    # -------------------------------------------------------------------------
    add_glob(
        candidates,
        r"ablation_trivial_baselines_outputs/summary/*",
        20,
        "trivial baseline aggregate summary",
    )

    add_glob(
        candidates,
        r"robustness_attribute_structure_outputs/summary/*",
        25,
        "robustness aggregate summary",
    )

    add_glob(
        candidates,
        r"kkt_anomaly_outputs/summary/*.csv",
        30,
        "KKT aggregate CSV summary",
    )
    add_glob(
        candidates,
        r"kkt_anomaly_outputs/summary/*.txt",
        31,
        "KKT aggregate text summary",
    )
    add_glob(
        candidates,
        r"kkt_anomaly_outputs/summary/*.json",
        32,
        "KKT aggregate JSON summary",
    )

    add_glob(
        candidates,
        r"kkt_anomaly_outputs/summary/FINAL_STATISTICS/*",
        33,
        "KKT final statistics",
        limit=20,
    )

    add_glob(
        candidates,
        r"kkt_anomaly_outputs/summary/PIPELINE_COMPLETION_AUDIT/*",
        34,
        "KKT pipeline completion audit",
        limit=20,
    )

    if INCLUDE_SUMMARY_FIGURES:
        add_glob(
            candidates,
            r"kkt_anomaly_outputs/summary/figures/*.pdf",
            35,
            "KKT summary figure PDF",
            limit=10,
        )
        add_glob(
            candidates,
            r"kkt_anomaly_outputs/summary/figures/*.png",
            36,
            "KKT summary figure PNG",
            limit=10,
        )

    # -------------------------------------------------------------------------
    # 5. Data audit outputs: useful and lightweight
    # -------------------------------------------------------------------------
    add_glob(
        candidates,
        r"kkt_data_audit_outputs/*.json",
        40,
        "KKT data audit JSON",
        limit=20,
    )
    add_glob(
        candidates,
        r"kkt_data_audit_outputs/*.txt",
        41,
        "KKT data audit text",
        limit=20,
    )
    add_glob(
        candidates,
        r"kkt_data_audit_outputs/*.csv",
        42,
        "KKT data audit CSV",
        limit=20,
    )
    add_glob(
        candidates,
        r"kkt_data_audit_outputs/samples/*",
        43,
        "KKT lightweight audit sample",
        limit=20,
    )

    # -------------------------------------------------------------------------
    # 6. Selected compact JSON results, not all hundreds
    # -------------------------------------------------------------------------
    if INCLUDE_SELECTED_JSON:
        # Keep only representative JSONs: full setting first, then a few ablations.
        add_glob(
            candidates,
            r"ablation_trivial_baselines_outputs/results_json/*__full__*.json",
            50,
            "selected full-baseline JSON result",
            limit=25,
        )
        add_glob(
            candidates,
            r"ablation_trivial_baselines_outputs/results_json/*__attr_only__*.json",
            51,
            "selected attr-only JSON result",
            limit=10,
        )
        add_glob(
            candidates,
            r"ablation_trivial_baselines_outputs/results_json/*__struct_only__*.json",
            52,
            "selected struct-only JSON result",
            limit=10,
        )
        add_glob(
            candidates,
            r"robustness_attribute_structure_outputs/results_json/*original*.json",
            53,
            "selected robustness original JSON result",
            limit=20,
        )

    # -------------------------------------------------------------------------
    # 7. Selected logs only; do not copy hundreds
    # -------------------------------------------------------------------------
    if INCLUDE_SELECTED_LOGS:
        add_exact(
            candidates,
            r"ablation_trivial_baselines_outputs\logs\MASTER_RUN_LOG.txt",
            60,
            "master ablation run log",
        )
        add_glob(
            candidates,
            r"ablation_trivial_baselines_outputs/logs/dataset_load_*.txt",
            61,
            "dataset loading log",
            limit=20,
        )
        add_glob(
            candidates,
            r"kkt_anomaly_outputs/logs/*.txt",
            62,
            "selected KKT log",
            limit=10,
        )
        add_glob(
            candidates,
            r"robustness_attribute_structure_outputs/logs/*.txt",
            63,
            "selected robustness log",
            limit=10,
        )

    candidates = deduplicate(candidates)
    candidates = sorted(candidates, key=lambda x: (x["priority"], x["sort_key"], x["size"]))

    return candidates


# =============================================================================
# MAIN
# =============================================================================

def main():
    if not SRC_ROOT.exists():
        raise FileNotFoundError(f"Source folder does not exist: {SRC_ROOT}")

    if DST_ROOT.exists():
        print(f"[INFO] Removing old destination: {DST_ROOT}")
        shutil.rmtree(DST_ROOT)

    DST_ROOT.mkdir(parents=True, exist_ok=True)

    # Generated files also count:
    # README.md, .gitignore, MANIFEST.json, MANIFEST_included_files.csv, MANIFEST_skipped_by_limit.csv
    generated_count = 5
    max_source_files = MAX_FILES - generated_count

    candidates = build_candidates()
    selected = candidates[:max_source_files]
    skipped_by_limit = candidates[max_source_files:]

    included_rows = []

    for item in selected:
        dst = DST_ROOT / item["rel"]
        copy_file(item["src"], dst)

        included_rows.append({
            "path": str(item["rel"]),
            "size_bytes": item["size"],
            "size_human": human_size(item["size"]),
            "priority": item["priority"],
            "reason": item["reason"],
        })

    write_text(DST_ROOT / "README.md", README_MD)
    write_text(DST_ROOT / ".gitignore", GITIGNORE)

    manifest = {
        "project": PROJECT_NAME,
        "created_at": datetime.now().isoformat(timespec="seconds"),
        "source_root": str(SRC_ROOT),
        "destination_root": str(DST_ROOT),
        "max_files": MAX_FILES,
        "max_file_mb": MAX_FILE_MB,
        "copied_source_files": len(selected),
        "generated_files": generated_count,
        "actual_file_count_expected": len(selected) + generated_count,
        "copied_total_size_bytes": sum(x["size"] for x in selected),
        "copied_total_size_human": human_size(sum(x["size"] for x in selected)),
        "included_files": included_rows,
        "skipped_by_limit_count": len(skipped_by_limit),
        "skipped_by_limit_preview": [
            {
                "path": str(x["rel"]),
                "size_human": human_size(x["size"]),
                "priority": x["priority"],
                "reason": x["reason"],
            }
            for x in skipped_by_limit[:200]
        ],
    }

    write_text(
        DST_ROOT / "MANIFEST.json",
        json.dumps(manifest, indent=2, ensure_ascii=False)
    )

    with open(DST_ROOT / "MANIFEST_included_files.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(
            f,
            fieldnames=["path", "size_bytes", "size_human", "priority", "reason"]
        )
        writer.writeheader()
        writer.writerows(included_rows)

    with open(DST_ROOT / "MANIFEST_skipped_by_limit.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(
            f,
            fieldnames=["path", "size_bytes", "size_human", "priority", "reason"]
        )
        writer.writeheader()
        for x in skipped_by_limit:
            writer.writerow({
                "path": str(x["rel"]),
                "size_bytes": x["size"],
                "size_human": human_size(x["size"]),
                "priority": x["priority"],
                "reason": x["reason"],
            })

    actual_files = [p for p in DST_ROOT.rglob("*") if p.is_file()]
    total_size = sum(safe_size(p) for p in actual_files)

    notebooks = [p for p in actual_files if p.suffix.lower() == ".ipynb"]
    py_files = [p for p in actual_files if p.suffix.lower() == ".py"]
    csv_files = [p for p in actual_files if p.suffix.lower() == ".csv"]
    json_files = [p for p in actual_files if p.suffix.lower() == ".json"]
    txt_files = [p for p in actual_files if p.suffix.lower() == ".txt"]

    print("=" * 100)
    print("DONE")
    print("=" * 100)
    print(f"Source        : {SRC_ROOT}")
    print(f"Destination   : {DST_ROOT}")
    print(f"Max files     : {MAX_FILES}")
    print(f"Actual files  : {len(actual_files)}")
    print(f"Repo size     : {human_size(total_size)}")
    print()
    print(f"Notebooks     : {len(notebooks)}")
    print(f"Python files  : {len(py_files)}")
    print(f"CSV files     : {len(csv_files)}")
    print(f"JSON files    : {len(json_files)}")
    print(f"TXT files     : {len(txt_files)}")
    print()

    if len(actual_files) <= MAX_FILES:
        print("[OK] File count is within the limit.")
    else:
        print("[WARNING] File count exceeded the limit.")

    print()
    print("Included notebooks:")
    for p in sorted(notebooks, key=lambda x: str(x).lower()):
        print(f"  - {p.relative_to(DST_ROOT)}")

    print()
    print("First included files:")
    for row in included_rows[:70]:
        print(f"  [{row['priority']:>2}] {row['path']}  ({row['size_human']}) -- {row['reason']}")

    print()
    print("Created:")
    print(f"  - {DST_ROOT / 'README.md'}")
    print(f"  - {DST_ROOT / '.gitignore'}")
    print(f"  - {DST_ROOT / 'MANIFEST.json'}")
    print(f"  - {DST_ROOT / 'MANIFEST_included_files.csv'}")
    print(f"  - {DST_ROOT / 'MANIFEST_skipped_by_limit.csv'}")

    print()
    print("Next commands:")
    print(f'  cd "{DST_ROOT}"')
    print("  git init")
    print("  git add .")
    print('  git commit -m "Initial lightweight ablation release"')


if __name__ == "__main__":
    main()
````
